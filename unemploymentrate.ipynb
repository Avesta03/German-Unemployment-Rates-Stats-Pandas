```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('1991-2021rates.csv', parse_dates=True)
```

```python
print(df.head())
```

```python
['year', 'unemployment rate']

small_df = df[['year','unemployment rate']]
```

```python
df.head()
```

```python
small_df.head()
```

```python
plt.plot(small_df['year'], small_df['unemployment rate'], color='red')
plt.title('German Unemployment Rate From 1991 to 2021')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

```python
plt.plot(df['year'], df['annual change'], color='blue')
plt.title('Change in Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Annual Change')
plt.show()
```

```python
import matplotlib.pyplot as plt

# create figure and axes for subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# plot first graph on first subplot
ax1.plot(small_df['year'], small_df['unemployment rate'], color='red')
ax1.set_title('German Unemployment Rate From 1991 to 2021')
ax1.set_xlabel('Year')
ax1.set_ylabel('Unemployment Rate (%)')

# plot second graph on second subplot
ax2.plot(df['year'], df['annual change'], color='blue')
ax2.set_title('Change in Unemployment Rate')
ax2.set_xlabel('Unemployment Rate (%)')
ax2.set_ylabel('Annual Change')

# adjust layout and display figure
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot(small_df['year'], small_df['unemployment rate'], color='red', label='Unemployment Rate')
plt.title('German Unemployment Rate From 1991 to 2021')
plt.xlabel('Year')

# Plotting second graph on the same figure
plt.plot(df['year'], df['annual change'], color='blue', label='Annual Change')
plt.title('Change in Unemployment Rate')
plt.xlabel('Year')

# Top right box
plt.legend()

# display figure
plt.show()
```
